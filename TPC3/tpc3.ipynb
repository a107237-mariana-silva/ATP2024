## TPC3: Jogo dos 21 fósforos

Desenvolve em Python o código necessário para o jogo dos 21 fósforos: 

### O jogo

* No início do jogo, há 21 fósforos;
* Cada jogador (computador ou utilizador), pode tirar 1, 2, 3 ou 4 fósforos quando for a sua vez de jogar;
* Os jogadores jogam alternadamente;
* **Quem tirar o último fósforo perde!**

### O programa 

* O jogo deverá ter dois modos: o jogador joga em primeiro lugar e o computador começa a jogar em segundo lugar e, no segundo modo, o computador começa em primeiro; 
* Quando o computador começa a jogar em segundo lugar, deve ganhar sempre o jogo;
* Quando o computador começa a jogar em primeiro lugar, se o utilizador cometer um erro de cálculo, o computador deverá passar para a posição de vencedor e ganhar o jogo.

```python
# Juntar os dois modos

from random import randint
fosforos = 21

print("""Olá! Vamos jogar ao jogo dos 21 fósforos.
No início do jogo, há 21 fósforos e cada jogador pode tirar entre 1 a 4 fósforos quando for a sua vez.
Os jogadores jogam alternadamente e quem tirar o último fósforo perde.
Existem dois modos: você começa (1) ou eu começo (2).
Vamos a isso?""")


while fosforos > 0:
      modo = int(input("Que modo quer jogar? (1/2)"))
      if modo == 1:
            print("Então você começa!")
            while fosforos > 0:
                  jog = int(input("Quantos quer retirar?"))
                  if (jog==1 or jog==2 or jog==3 or jog==4):
                        fosforos = fosforos - jog
                        print(f"Ainda faltam {fosforos} fósforos.")
                        if fosforos%5 == 0 and fosforos > 5:
                              print("Eu vou retirar 4.")
                              fosforos = fosforos - 4
                              print(f"Ainda faltam {fosforos} fósforos.")
                        elif fosforos%5 != 0 and fosforos > 5:
                              comp1 = fosforos%5 - 1
                              print(f"Eu vou retirar {comp1}.")
                              fosforos = fosforos - comp1
                              print(f"Ainda faltam {fosforos} fósforos.")
                        elif fosforos <= 5:
                              comp2 = fosforos - 1
                              print(f"Eu vou jogar {comp2}.")
                              fosforos = fosforos - comp2
                              print(f"Ainda faltam {fosforos} fósforos.")
                              jog2 = int(input("Quantos quer retirar?"))
                              if jog2 == 1:
                                   print("Já não há mais fósforos... Retirou o último. Eu ganhei!")
                                   fosforos = 0
                              else:
                                   while jog2 != 1:
                                          print("Já não há assim tantos fósforos!")
                                          jog2 = int(input("Quantos quer retirar?"))
                                          if jog2 == 1:
                                                print("Já não há mais fósforos... Retirou o último. Eu ganhei!")
                                                fosforos = 0
                  else:
                        print("Resposta incorreta...")
                        print("Por favor, introduza um numero inteiro entre 1 e 4 inclusive.")
      elif modo == 2:
           print("Então começo eu!")
           comp3 = randint(1,4)
           print(f"Eu vou retirar {comp3}.")
           fosforos = fosforos - comp3
           print(f"Ainda faltam {fosforos} fósforos.")
           while fosforos >0:
                 jog3 = int(input("Quantos quer retirar?"))
                 if (jog3==1 or jog3==2 or jog3==3 or jog3==4):
                       fosforos = fosforos - jog3
                       print(f"Ainda faltam {fosforos} fósforos.")
                       if fosforos%5 == 0 and fosforos >= 6:
                            alea = randint(1,4)
                            print(f"Eu vou retirar {alea}.")
                            fosforos = fosforos - alea
                            print(f"Ainda faltam {fosforos} fósforos.")
                       elif fosforos%5 != 0 and fosforos >= 6:
                            resto = (fosforos%5)
                            print(f"Eu vou retirar {resto}.")
                            fosforos = fosforos - resto
                            print(f"Ainda faltam {fosforos} fósforos.")
                       elif fosforos < 6 and fosforos > 1:
                            print(f"Eu vou retirar {fosforos - 1}.")
                            fosforos = 1
                            print(f"Ainda falta 1 fósforo.")
                            jog4 = int(input("Quantos quer retirar?"))
                            if jog4 == 1:
                                  print("Já não há mais fósforos... Retirou o último. Eu ganhei!")
                                  fosforos = 0
                            else:
                                  while jog4 != 1:
                                          print("Já não há assim tantos fósforos!")
                                          jog4 = int(input("Quantos quer retirar?"))
                                          if jog4 == 1:
                                                print("Já não há mais fósforos... Retirou o último. Eu ganhei!")
                                                fosforos = 0
                       elif fosforos == 1:
                            print("Vou ter de retirar 1.")
                            print("Já não há mais fósforos... Você ganhou!")
                            fosforos = 0
                 else:
                       print("Resposta incorreta...")
                       print("Por favor, introduza um numero inteiro entre 1 e 4 inclusive.")
      else:
            print("Só pode escolher entre dois modos.")
```
