### TPC6: Aplicação para gestão de alunos

Considere que o modelo do aluno e da turma têm a seguinte estrutura:

`aluno = (nome, id, [notaTPC, notaProj, notaTeste])`

`turma = [aluno]`

* Cria uma aplicação que coloca no monitor o seguinte menu de operações:
    - 1: Criar uma turma;
    - 2: Inserir um aluno na turma;
    - 3: Listar a turma;
    - 4: Consultar um aluno por id;
    - 8: Guardar a turma em ficheiro;
    - 9: Carregar uma turma dum ficheiro;
    - 0: Sair da aplicação
* No fim de executar a operação selecionada, a aplicação deverá colocar novamente o menu e pedir ao utilizador a opção para continuar;
* Utiliza a tua aplicação para criar uma turma com 5 alunos.

```python

# Função auxiliar
# True -> Já existe dentro da turma
# False -> Ainda não existe dentro da turma
def existeAluno(turma,id):
    cond = False
    if len(turma) > 0:
        for aluno in turma:
            if aluno[1] == id:
                cond = True
    else:
        cond = False
    return cond


def inserirAluno (turma,id):
    nome = input("Introduza o nome do aluno:")
    notaTPC = input("Introduza a nota final dos TPCs do aluno:")
    notaProj = input("Introduza a nota do Projeto do aluno:")
    notaTeste = input("Introduza a nota do Teste do aluno:")
    notas = [int(notaTPC), int(notaProj),int(notaTeste)]
    aluno = (nome,id,notas)
    turma.append(aluno)
    return turma


def listaTurma (turma):
    if len(turma) > 0:
        print(f"- - - - - Lista de Alunos - - - - -")
        for aluno in turma:
            print(f"Nome: {aluno[0]} | id : {aluno[1]} | Lista de Notas: {aluno[2]}")
    else:
        print("Pedimos desculpa, mas a sua turma não tem nenhum aluno.")
    return "\n"


def consultaAluno (turma, id):
    if len(turma) > 0:
        if not existeAluno(turma,id):
            print(f"Pedimos desculpa, mas nenhum aluno dentro desta turma possui esse id.")
        else:
            for aluno in turma:
                if aluno[1] == id:
                    print(f"Aluno: {aluno[0]} | id : {aluno[1]} | Lista de Notas: {aluno[2]}")
    else:
        print("A sua turma não tem nenhum aluno.")
    return "\n" 


# Separar alunos -> \n (por linha)
# Separar as secções -> |
# Separar as notas -> &
def guardarT_Ficheiro(turma):
    file = open("turma", "w")
    for aluno in turma:
        nome,id,notas = aluno[0],aluno[1],aluno[2]
        notaTPC, notaProj, notaTeste = notas
        for nota in notas:
            n = (str(notaTPC) + " & " + str(notaProj)+ " & " + str(notaTeste))
        file.write(f"{nome} | {id} | {n}\n")
    file.close()
    return ("O seu ficheiro da turma foi guardado.")


def carregarT_Ficheiro ():
    turma = []
    file = open("turma","r")
    for line in file:
        secções = line.split(" | ")
        nome,id,notas = secções[0],secções[1],secções[2]
        nota = notas.split(" & ")
        notaTPC,notaProj,notaTeste = nota[0], nota[1], nota[2]
        listaNotas = [int(notaTPC),int(notaProj),int(notaTeste)]
        aluno = (nome,id,listaNotas)
        turma.append(aluno)
    file.close()
    return turma


def criaTurma (turma):
    turma = []
    return turma


def sair():
    return ("A aplicação irá encerrar.")
        

def menu():
    print(""" - - - - - MENU - - - - - 
1 - Criar Turma
2 - Acrescentar Aluno
3 - Lista da Turma
4 - Consultar Aluno
5 - Guardar num Ficheiro
6 - Carregar o Ficheiro
0 - Sair""")
    

aluno1 = ("Manuel da Silva","a134967",[14,15,11])
aluno2 = ("João Maia Carvalho","a134551",[12,14,10])
aluno3 = ("Lisa Batista Campos","a135082",[13,14,13])
aluno4 = ("Marco Saraiva","a134811",[16,17,12])
aluno5 = ("Emilie Cunha Santos","a135519",[11,13,10])
turma = [aluno1,aluno2,aluno3,aluno4,aluno5]
```

```python
turma = []
cond = True
while cond:
    menu()
    opcao = int(input(("Introduza a opção pretendida:")))
    if opcao == 1:
        print("A sua turma foi criada.")
        print(criaTurma(turma))
    elif opcao == 2:
        id = input("Introduza o id do aluno que quer inserir:")
        if not existeAluno(turma,id):
            print(inserirAluno(turma,id))
        else:
            print("Nenhum aluno pode ter o mesmo id.")
    elif opcao == 3:
        print(listaTurma(turma))
    elif opcao == 4:
        id = (input("Introduza o id do aluno que quer consultar:"))
        print(consultaAluno(turma,id))
    elif opcao == 5:
        print(guardarT_Ficheiro(turma))
    elif opcao == 6:
        print(carregarT_Ficheiro())
    elif opcao == 0:
        print(sair())
        cond = False
    else:
        print("Por favor, introduza um numero entre 0 e 7.")
```
