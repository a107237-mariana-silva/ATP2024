### TPC5: Aplicação para Gerir um Cinema

Suponha que está a desenvolver uma aplicacão para gestão de um conjunto de salas de cinema de um centro comercial. 
Nesse centro comercial existem algumas salas de cinema (que poderão estar a exibir filmes ou não), cada sala tem uma determinada 
lotação, uma lista com a referência dos bilhetes vendidos (lugares ocupados; cada lugar é identificado por um número inteiro), e cada sala tem um filme associado.

Considera a seguinte sugestão para o modelo dos cinemas:
```
Cinema = [Sala]
Sala = [nlugares, Vendidos, filme]
nlugares = Int
filme = String 
Vendidos = [Int]
```
  
Que poderá ser usado num programa da seguinte forma:
```
sala1 = (150, [], "Twilight")
sala2 = (200, [], "Hannibal")
cinema1 = []
...
cinema1 = inserirSala(cinema1,sala1)
cinema1 = inserirSala(cinema1,sala2)
...
listar(cinema1)
...

if(disponivel(cinema1, "Twilight", 17 )):
  cinema1 = vendebilhete(cinema1, "Twilight", 17 )
...
listardisponibilidades(cinema1)
...
```

Especifique as funções utilizadas no exemplo:

1. `listar( cinema )` - que lista no monitor todos os filmes que estão em exibição nas salas do cinema passado como argumento;
2. `disponivel( cinema, filme, lugar )` - que dá como resultado **False** se o lugar lugar já estiver ocupado na sala onde o filme está a ser exibido e dará como resultado **True** se o inverso acontecer;
3. `vendebilhete( cinema, filme, lugar )` - que dá como resultado um novo cinema resultante de acrescentar o lugar à lista dos lugares ocupados, na sala onde está a ser exibido o filme;
4. `listardisponibilidades( cinema )` - que, para um dado cinema, lista no monitor para cada sala, o filme que está a ser exibido e o total de lugares disponíveis nessa sala (número de lugares na sala menos o número de lugares ocupados);
5. `inserirSala( cinema, sala )` - que acrescenta uma sala nova a um cinema (devendo verificar se a sala já existe);
6. Acrescente todas as outras funcionalidades que achar necessárias;
7. À semelhança do exercício 3, construa uma aplicação com um menu de interface para as operações.

````python
def listarCinema (cinema):
    filmes = []
    for sala in cinema:
        filmes.append(sala[2])
    return (f"Filmes em exibição: {filmes}.")


def disponivel (cinema,filme,lugar):
    cond = True
    for sala in cinema:
        if filme in sala[2]:
            vendidos = sala[1]
            if lugar in vendidos:
                cond = False
    return cond


def vendeBilhete (cinema,filme,lugar):
    for sala in cinema:
        if sala[2] == filme:
            if lugar <= sala[0]:  
                vendidos = sala[1]
                vendidos.append(lugar)
                print("O lugar foi vendido com sucesso.")
            elif lugar > sala[0]:
                print("Esse lugar não existe nessa sala.")
    return cinema


def listarDisponibilidades (cinema):
    for sala in cinema:
        disponiveis = sala[0] - len(sala[1])
        print((f"Para o filme: {sala[2]}, ha {disponiveis} lugares disponiveis."))
    return cinema


# Esta função serve para ver se existe uma sala com o mesmo filme a ser exibido
# True -> Está
# False -> Não está
def existeSala(cinema,sala):
    cond = False
    for s in cinema:
        if s[2] == sala[2]:
            cond = True
    return cond


def inserirSala (cinema,sala):
    if not existeSala(cinema,sala):
            cinema.append(sala)
    else:
        print("Essa sala já existe.")
    return cinema


# Esta função vai servir a retirar um certo lugar da lista de Vendidos porque é reembolsado
def reembolso (cinema,filme,lugar):
    for sala in cinema:
        if filme in sala[2]:
            if lugar in sala[1]:
                vendidos = sala[1]
                vendidos.remove(lugar)
                print("Esse lugar foi reembolsado com sucesso.")
            else:
                print("Esse bilhete não está vendido.")
    return cinema


# Esta função vai nos mostrar que lugares estão ocupados dentro de uma certa sala
def vendidos (cinema,filme):
    for sala in cinema:
        if filme == sala[2]:
            print (f"Filme em exibição: {sala[2]} | Lugares vendidos: {sala[1]}")
    return cinema


# Função Sair
def sair():
    return ("Muito obrigada e até à proxima.")


def menu():
    print(""" - - - - - MENU - - - - - 
1 - Lista dos Filmes
2 - Acrescentar uma Sala
3 - Lista de Disponibilidades
4 - Bilhete Disponivel
5 - Venda de um Bilhete
6 - Lista de Vendidos
7 - Reembolso
0 - Sair""")
    

# Verificar se funciona:

sala1 = (150,[], "Twilight")
sala2 = (200,[], "Hannibal")
cinema = []
cinema = inserirSala(cinema,sala1)
cinema = inserirSala(cinema,sala2)

print("1")
print(cinema)

print(listarCinema(cinema))

print("2")
print(reembolso(cinema,"Twilight",17))
print(vendidos(cinema,"Twilight"))

print("3")
print(disponivel(cinema,"Twilight",17))
if(disponivel(cinema,"Twilight",17)):
    print(vendeBilhete(cinema,"Twilight",17))
print(disponivel(cinema,"Twilight",17))
print(vendidos(cinema,"Twilight"))

print("4")
if(disponivel(cinema,"Twilight",170)):
    print(vendeBilhete(cinema,"Twilight",170))

print("5")
print(listarDisponibilidades(cinema))

print("6")
print(reembolso(cinema,"Twilight",17))
print(listarDisponibilidades(cinema))

print("7")
print(cinema)
print(sair())
````
